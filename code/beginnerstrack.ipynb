```python
# Imports
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
import matplotlib.pyplot as plt
from scipy import stats
from datetime import date

warnings.filterwarnings("ignore")


# Daten werden eingelesen
crimes_dataset = pd.read_csv(r"Crimes_Dataset.csv")
suspects_dataset = pd.read_csv(r"Suspects_Dataset.csv")
```

Data Exploration

```python
crimes_dataset = crimes_dataset.applymap(lambda x: x.lower() if isinstance(x, str) else x)   # Alle Zeichenfolgen in den DataFrame in Kleinbuchstaben umwandeln
crimes_dataset
```

```python
suspects_dataset = suspects_dataset.applymap(lambda x: x.lower() if isinstance(x, str) else x)    # Alle Zeichenfolgen in den DataFrame in Kleinbuchstaben umwandeln
suspects_dataset
```

```python
print(suspects_dataset.isnull().sum())                       #welche Zeilen haben keinen Eintrag
print("___________________________________________")
crimes_dataset.isnull().sum()                               #welche Zeilen haben keinen Eintrag
```

Data cleaning

```python
#search for dublicates
print(crimes_dataset.duplicated().sum(), "\n\n---------------\n\n", suspects_dataset.duplicated().sum())

#if dublicates > 0 --> unvomment following line for drop the dublicate rows
#crime_dataset = crime_dataset.drop_dublicates()
#suspects_dataset = suspects_dataset.drop_duplicates()
```

```python
#Rename a column

crimes_dataset.rename(columns={
                             "Monster involved":"Monster_involved",
                              "Days of Investigation": "Days_of_Investigation",
                              "Crime Type":"Crime_Type",
                              "Crime Weapon": "Crime_Weapon",
                              "Time of Day": "Time_of_Day",
                              "Evidence Found":"Evidence_Found"
                                }, inplace=True)      #inplace=True da sonst zwei Data frames erstellt werden, so muss der neue Data frame nicht umbenannt und neu eingespeichert werden

suspects_dataset.rename(columns={
    
    "Criminal record":"Criminal_record",
    "Height in cm": "Height_in_cm",
    "Speed Level": "Speed_Level",
    'Strength Level': 'Strength_Level',
    'Favorite Food': 'Favorite_Food'
}, inplace=True)       


print(crimes_dataset.head())

print ("-------------------")

suspects_dataset.head()

#Kontrolle ob umbenannt wurde
```

```python
#converting data types#

print(suspects_dataset.info())                                                     
print("-----------------------------------------------------------------")
print(crimes_dataset.info()) 


crimes_dataset["Date"] = pd.to_datetime(crimes_dataset["Date"], format= "%Y-%m-%d")
```

```python
#handling outliers

# Calculate Z-scores for 'crimes_dataset'
z_scores1 = stats.zscore(crimes_dataset['Index_Crimes'])
z_scores2 = stats.zscore(crimes_dataset['Days_of_Investigation'])

#Calculate Z-scores for "suspects_dataset"
z_scores3 = stats.zscore(suspects_dataset['Index_Monster'])
z_scores4 = stats.zscore(suspects_dataset['Age'])
z_scores5 = stats.zscore(suspects_dataset['Height_in_cm'])
z_scores6 = stats.zscore(suspects_dataset['Speed_Level'])
z_scores7 = stats.zscore(suspects_dataset['Strength_Level'])

print(z_scores1)
print(z_scores2)
print(z_scores3)
print(z_scores4)
print(z_scores5)
print(z_scores6)
print(z_scores7)


# Remove outliers
#crimes_dataset_no_outliers = crimes_dataset[(np.abs(z_scores1) < 3)]
#suspects_dataset_no_outliers = suspects_dataset[(np.abs(z_scores) < 3)]
#haben keine Abweichung von über 3
```
